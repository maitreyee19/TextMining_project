```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib
```

```python
df= pd.read_excel("F:/ALL SUB/SEM 2/Text mining/Project/tweet1.xlsx")
df.head(5)
```

```python
tweets=df.loc[(df['date'] >= "2020-03-01") & (df['date'] <= "2020-03-31"),'Tweets' ]
all_tweets= df['Tweets']
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words= stopwords.words('english')
newStopWords = ['market','stock','stock market','&amp','#stockmarket','-','us','like','time','&amp;','market.','stocks','would']
stop_words.extend(newStopWords)


```

```python
output ={}
for i in all_tweets:
#     print(i)
    words= str(i).lower().split()
    for word in words:
        if word in stop_words:
            continue
        else:
            if word not in output.keys():
                output[word] = 1
            output[word] +=1
# print(output)
```

```python
a = sorted(output.items(), key=lambda x: x[1],reverse=True)    
print(a)
```

```python
background_color = "#101010"
height = 720
width = 1080

word_cloud = WordCloud(
    background_color=background_color,
    width=width,
    height=height
)

word_cloud.generate_from_frequencies(output)
word_cloud.to_file('Combine_wordcloud.png')
```
